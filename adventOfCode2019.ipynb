```python
with open("data-2019/inputday20.txt", 'r') as f:
    day20data = [x for x in f.read().split("\n") if x]
```

```python
import numpy as np
import collections, itertools
```

```python
print(list(set(len(x) for x in day20data)), len(day20data))
xlim, ylim = 135, 127
```

```python
isLetter = lambda x: ord(x) >= ord("A") and ord(x) <= ord("Z")
```

```python
maze = collections.defaultdict(list) # with (x, y) -> [(x, y)]
portals = collections.defaultdict(list) # with ("AB") -> [(x, y)]
for i, j in itertools.product(range(ylim), range(xlim)):
    if day20data[i][j] == ".":
        # only need to look down and right
        if day20data[i+1][j] == ".":
            maze[(i, j)].append((i+1, j))
            maze[(i+1, j)].append((i, j))
        if day20data[i][j+1] == ".":
            maze[(i, j)].append((i, j+1))
            maze[(i, j+1)].append((i, j))
        if isLetter(day20data[i+1][j]) and isLetter(day20data[i+2][j]):
            portals[day20data[i+1][j] + day20data[i+2][j]].append((i, j))
        if isLetter(day20data[i][j+1]) and isLetter(day20data[i][j+2]):
            portals[day20data[i][j+1] + day20data[i][j+2]].append((i, j))
    try:
        if isLetter(day20data[i][j]) and isLetter(day20data[i+1][j]) and day20data[i+2][j] == ".":
            portals[day20data[i][j] + day20data[i+1][j]].append((i+2, j))
        if isLetter(day20data[i][j]) and isLetter(day20data[i][j+1]) and day20data[i][j+2] == ".":
            portals[day20data[i][j] + day20data[i][j+1]].append((i, j+2))
    except IndexError:
        pass
            
```

## Make paths
Only consider nodes and paths from there

```python
paths = collections.defaultdict(list) # [(x, y), (x, y), weight]
mazeKeys = [x for x,y in maze.items() if len(y) > 2]
nodes = mazeKeys.copy()# [(x, y), ]
visited = []
for start in mazeKeys:
    visited.append(start)
    #print(start, maze[start])
    for branch in maze[start]:
        if branch not in visited:
            weight = 0
            prev = s = start
            ds = [branch, branch]
            while len(ds) == 2:
                #print(weight, ds, prev)
                s, prev = [x for x in ds if x != prev][0], s
                ds = maze[s]
                #print(s, ds)
                if len(ds) < 3:
                    visited.append(s)
                weight += 1
            # hit either end point or branch point, doesn't matter
            paths[start].append((s, weight))
            paths[s].append((start, weight))
            if len(ds) == 1:
                nodes.append(s)
```

## Remove dead ends

```python
ENTRY, EXIT = portals['AA'][0], portals['ZZ'][0]
```

```python
## Let's remove dead ends:
nodeCount = len(nodes)
shouldContinue = True
while shouldContinue:
    print(f"Nodecount {nodeCount}")
    allNodes = nodes.copy()
    for node in allNodes:
        if node in [ENTRY, EXIT]:
            continue
        pn = paths[node]
        if len(pn) == 1:
            nodes.remove(node)
            del paths[node]
            paths[pn[0]] = [(d, w) for d, w in paths[pn[0]] if d != node]
            if len(paths[pn[0]]) == 0:
                del paths[pn[0]]
    shouldContinue = len(nodes) != nodeCount
    nodeCount = len(nodes)
```

## Breadth-first search through the paths

```python
# OK, lets search for the path:
bestpath = {ENTRY : (0, ENTRY)} # {(x,y) : (cost, prevNode)}
tovisit = [ENTRY]
while len(tovisit) != 0:
    node = tovisit.pop()
    for dest, w in paths[node]:
        destCost = bestpath[node][0] + w
        if dest in bestpath:
            if bestpath[dest][0] > destCost:
                bestpath[dest] = (destCost, node)
                tovisit.append(dest)
        else:
            bestpath[dest] = (destCost, node)
            tovisit.append(dest)
```

```python
def getPath(node, entry, bestpath):
    nodes = [node]
    while nodes[-1] != entry:
        nodes.append(bestpath[nodes[-1]][1])
    return nodes
```

```python
path = getPath(EXIT, ENTRY, bestpathN)[::-1]
```

```python
def makeMaze2(maze, portals, depth):
    mazeNew = collections.defaultdict(list)
    for i in range(depth):
        for (x, y), v in maze.items():
            mazeNew[(x, y, i)] = [(x1, y1, i) for x1, y1 in v]
    for i in range(depth):
        for portal in portals.values():
            if len(portal) == 2:
                inner = [(x,y) for x,y in portal if x in range(36, 91) and y in range(36, 99)][0]
                outer = [x for x in portal if x != inner][0]
                mazeNew[(inner[0], inner[1], i)].append((outer[0], outer[1], i+1))
                mazeNew[(outer[0], outer[1], i+1)].append((inner[0], inner[1], i))
    return mazeNew
```
